````python
import numpy as np
import matplotlib.pyplot as plt

def fncBeale(x, fparam=None):
    x1,x2 = x
    x22   = x2*x2
    x23   = x22*x2
    return (1.5-x1 + x1*x2)**2 + (2.25 - x1 + x1*x22)**2 + (2.625 - x1 + x1*x23)**2

def gradBeale(x, fparam=None):
    x1,x2 = x
    x22 = x2*x2
    x23 = x22*x2
    t1  = 1.5-x1 + x1*x2
    t2  = 2.25 - x1 + x1*x22 
    t3  = 2.625 - x1 + x1*x23 
    g1  = 2*t1*(x2-1) + 2*t2*(x22-1) + 2*t3*(x23-1)
    g2  = 2*t1*x1     + 4*t2*x1*x2   + 6*t3*x1*x22
    return np.array([g1,g2])


# Función objetivo
def fncHimmelblau(x, fparam=None):
    return (x[0]**2 + x[1] - 11)**2 + (x[0] + x[1]**2 - 7)**2;

# Gradiente de la función objetivo
def gradHimmelblau(x, fparam=None):
    gx = 4*x[0]*(x[0]**2 + x[1] - 11) + 2*(x[0] + x[1]**2 - 7)
    gy = 2*(x[0]**2 + x[1] - 11) + 4*x[1]*(x[0] + x[1]**2 - 7)
    return np.array([gx, gy])


def fncRosenbrock(x, fparam=None):
    n = len(x)
    dsum = 0.0
    for i in range(n-1):
        dsum += 100*(x[i+1] - x[i]**2)**2 + (1-x[i])**2
    return dsum

def gradRosenbrock(x, fparam=None):
    n = len(x)
    g = np.zeros(n)
    g[0] = -400*x[0]*(x[1] - x[0]**2) - 2*(1-x[0])
    for i in range(1,n-1):
        g[i] = -400*x[i]*(x[i+1] - x[i]**2) - 2*(1-x[i]) + 200*(x[i] - x[i-1]**2)
    g[n-1] = 200*(x[n-1] - x[n-2]**2)
    return g


def testGrad(f, gradf, h, x0):
    n  = len(x0)
    I  = np.eye(n)
    f0 = f(x0)
    gh = np.zeros(n)
    for i in range(n):
        gh[i] = (f(x0+h*I[i])-f0)/h
    g  = gradf(x0)
    return np.linalg.norm(g-gh)/np.linalg.norm(g)

h  = 0.000001
nt = 10
ve = np.zeros(nt)
for it in range(nt):
    x0 = np.random.randn(2)
    ve[it] = testGrad(fncBeale, gradBeale, h, x0)
print('Errores: ', np.min(ve), np.mean(ve), np.max(ve))

for it in range(nt):
    x0 = np.random.randn(2)
    ve[it] = testGrad(fncHimmelblau, gradHimmelblau, h, x0)
print('Errores: ', np.min(ve), np.mean(ve), np.max(ve))


for it in range(nt):
    x0 = np.random.randn(2)
    ve[it] = testGrad(fncRosenbrock, gradRosenbrock, h, x0)
print('Errores: ', np.min(ve), np.mean(ve), np.max(ve))
````

````python
import matplotlib.pyplot as plt
import numpy as np


def plot2DFnc(fncf, xleft, xright, ybottom, ytop, levels=None, mpts=None):
    # Crea una discretización uniforme del intervalo [xleft, xright]
    ax = np.linspace(xleft, xright, 250)
    # Crea una discretización uniforme del intervalo [ybottom, ytop]
    ay = np.linspace(ybottom, ytop, 250)
    # La matriz mX que tiene las abscisas 
    mX, mY = np.meshgrid(ax, ay)
    # Se crea el arreglo mZ con los valores de la función en cada nodo
    vz = np.zeros( len(ax)*len(ay) )
    for i,xy in enumerate(zip(mX.flatten(), mY.flatten())):
        vz[i] = fncf(xy)
    mZ = vz.reshape( len(ay), len(ax) ) 
    if levels is None:
        fig = plt.figure(figsize=(6,4))
        im = plt.imshow(mZ, cmap='coolwarm', extent=(xleft, xright, ybottom, ytop), 
                        interpolation='bilinear', origin='lower') 
        plt.colorbar(im)
    else:
        fig, ax = plt.subplots()
        CS = ax.contour(mX, mY, mZ, levels, cmap='Wistia')
    if mpts is not None:
        plt.plot(mpts[:,0], mpts[:,1], 'k')
    plt.title(r'$f(x,y)$')


def backtracking(fncf, xk, fk, gk, pk, aini, rho, c, MaxIter):
    dft = c*np.vdot(gk, pk)
    a   = aini
    for it in range(MaxIter):
        if fncf(xk + a*pk) <= (fk + dft*a):
            return a,it,True
        a *= rho
    return a,it,False


# Descenso maximo con busqueda exacta
def descensoMaximoBacktracking(fnc, grad, x0, N, tau, fparams, aini, rho, c1, Nbk):
    n    = len(x0)
    xk   = x0
    bres = False
    fk   = fnc(xk, fparams)
    gk   = grad(xk, fparams)
    if n==2:
        mpt = np.zeros((N+1,5))
        mpt[0,:] = x0[0],x0[1],1,0,fk
    for k in range(N):
        pk = -gk
        #ak = optimize.golden(fncPhi, args=(fnc, xk, pk), brack=(0, 0.5))
        ak,itbk,bresbk = backtracking(fnc, xk, fk, gk, pk, aini, rho, c1, Nbk)
        if not bresbk:
            print('Insuficientes iteraciones Backtracking', ak)
        xkp1 = xk + ak*pk
        fkp1 = fnc(xkp1, fparams)
        if n==2:
            mpt[k+1,:] = xkp1[0],xkp1[1],ak,itbk,fkp1
        if np.linalg.norm(xk-xkp1)<tau:
            bres = True
            break
        xk = xkp1
        fk = fkp1
        gk = grad(xk, fparams)
    if n==2:
        mpt = mpt[:(k+1),:]
    else:
        mpt = None
    return xk, fk, gk, k, bres, mpt
````

````python
N    = 30000
Nb   = 600
c1   = 0.1
rho  = 0.9
aini = 1.0

x0  = np.array([2.,3.])
x0  = np.array([2.,4.])  # No converge
n   = len(x0)
tau = np.sqrt(n) * np.finfo(float).eps**(1/2)

xk, fk, gk, k, bres, mpt = descensoMaximoBacktracking(fncBeale, gradBeale, x0, N, 
                                                      tau, [], aini, rho, c1, Nb)
np.set_printoptions(precision=4)
print('Num.Iter = ', k)
print('xk       = ', xk)
print('f(xk)    = ', fk)
print('||gk||   = ', np.linalg.norm(gk))
print('ak median= ', np.mean(mpt[:,2]))
print('it median= ', np.mean(mpt[:,3]))
print('bres     = ', bres)  

plot2DFnc(fncBeale, xleft=-4.5, xright=4.5, ybottom=-4.5, ytop=4.5, 
          levels=[0.5, 5, 15, 50, 150, 500, 1000, 2500.0, 5000.0], mpts=mpt)
````

````python

def backtracking(fncf, xk, fk, gk, pk, aini, rho, c, MaxIter):
    dft = c*np.vdot(gk, pk)
    a   = aini
    for it in range(MaxIter):
        if fncf(xk + a*pk) <= (fk + dft*a):
            return a,it,True
        a *= rho
    return a,it,False


# Descenso maximo con busqueda exacta
def descensoMaximoBacktracking(fnc, grad, x0, N, tau, fparams, aini, rho, c1, Nbk):
    n    = len(x0)
    xk   = x0
    bres = False
    fk   = fnc(xk, fparams)
    gk   = grad(xk, fparams)
    gpkm1= np.vdot(gk, -gk)
    akm1 = aini
    if n==2:
        mpt = np.zeros((N+1,8))
        mpt[0,:] = x0[0],x0[1],1,0,fk,1,1,1
    for k in range(N):
        pk   = -gk
        gpk  = np.vdot(gk, pk)
        aki  = akm1*np.min([100, gpkm1/gpk])
        #ak = optimize.golden(fncPhi, args=(fnc, xk, pk), brack=(0, 0.5))
        ak,itbk,bresbk = backtracking(fnc, xk, fk, gk, pk, aki, rho, c1, Nbk)
        #print('aki=', aki, ak,itbk,bresbk)
        if not bresbk:
            print('Insuficientes iteraciones Backtracking', ak)
        xkp1 = xk + ak*pk
        fkp1 = fnc(xkp1, fparams)
        if n==2:
            mpt[k+1,:] = xkp1[0],xkp1[1],ak,itbk,fkp1,aki,gpkm1,gpk
        if np.linalg.norm(xk-xkp1)<tau:
            bres = True
            break
        xk = xkp1
        fk = fkp1
        gk = grad(xk, fparams)
        akm1  = ak 
        gpkm1 = gpk
    if n==2:
        mpt = mpt[:(k+1),:]
    else:
        mpt = None
    return xk, fk, gk, k, bres, mpt
````

````python
N    = 30000
Nb   = 600
c1   = 0.1
rho  = 0.9
aini = 1.0

x0  = np.array([2.,3.])
x0  = np.array([2.,4.])  # No converge
n   = len(x0)
tau = np.sqrt(n) * np.finfo(float).eps**(1/2)

xk, fk, gk, k, bres, mpt = descensoMaximoBacktracking(fncBeale, gradBeale, x0, N, 
                                                      tau, [], aini, rho, c1, Nb)
np.set_printoptions(precision=4)
print('Num.Iter = ', k)
print('xk       = ', xk)
print('f(xk)    = ', fk)
print('||gk||   = ', np.linalg.norm(gk))
print('ak median= ', np.median(mpt[:,2]))
print('it median= ', np.median(mpt[:,3]))
print('bres     = ', bres)  

plot2DFnc(fncBeale, xleft=-4.5, xright=4.5, ybottom=-4.5, ytop=4.5, 
          levels=[0.5, 5, 15, 50, 150, 500, 1000, 2500.0, 5000.0], mpts=mpt)
````

````python
def golden_search(fncf, xl, xu, tau1, N, fparam=None):
    '''
        Metodo de la seccion dorada
    '''
    rho  = (np.sqrt(5) - 1) / 2
    for k in range(N):
        h    = xu - xl
        b    = rho*h
        x1   = xu - b
        x3   = xl + b
        fxl  = fncf(xl, fparam)
        fxu  = fncf(xu, fparam)
        fx1  = fncf(x1, fparam)
        fx3  = fncf(x3, fparam)
        if fx1 < fx3:
            xu  = x3
            xk  = x1
            fxk = fx1
        else:
            xl  = x1
            xk  = x3
            fxk = fx3
        #if np.abs(fx3 - fx1)<tau2 or h<tau1:
        if  (xu-xl)<tau1:
            return xk, fxk, xl, xu, k, True
    return xk, fxk, xl, xu, k, False



def descensoPorCoordenadas(fncf, gradf, x0, N, tau, xl, xu, Ngs, fparams=None):
    n   = len(x0)
    xk  = x0
    res = 0
    I   = np.eye(len(x0))
    fk  = fncf(xk, fparams)
    if gradf is not None:
        gk    = gradf(xk, fparams)
        gpkm1 = np.vdot(gk, -gk)
    mpt = np.zeros(((N+1)*len(x0),3))
    mpt[0,:] = x0[0], x0[1], fk
    mc  = 1
    for k in range(N):
        xkp1 = xk
        for i in range(2):
            pk   = I[i,:]
            fphi = lambda a,b: fncf(xkp1 + a*pk, b)
            ai,_,_,_,_,_ = golden_search(fphi, xl, xu, tau, Ngs, fparam=fparams)
            #ai = optimize.golden(fncPhi, args=(fnc, xkp1, di), brack=(-1, 1))
            xkp1 = xkp1 + ai*pk
            fkp1 = fncf(xkp1, fparams)
            mpt[mc,:] = xkp1[0], xkp1[1], fkp1
            mc  += 1
        if np.linalg.norm(xk-xkp1)<tau:
            res = 1
            break
        xk = xkp1
        fk = fkp1
    mpt = mpt[:mc,:]
    return xk, fk, k, res, mpt
````

````python
N   = 30000
Ngs = 500

x0  = np.array([2.,4.])  # No converge
x0  = np.array([2.,3.])
n   = len(x0)
tau = np.sqrt(n) * np.finfo(float).eps**(1/2)

print('Punto inicial x0=', x0)
print('f(x0) = ', fncBeale(x0))
xk, fk, k, res, mpt = descensoPorCoordenadas(fncBeale, x0, N, tau, -1, 1, Ngs, None)
if res==1:
    print('\nEl algoritmo converge en %d ciclos' % k)
else:
    print('\nEl algoritmo No converge en %d ciclos' % k)
print('xk = ', xk)
print('fk = ', fk)

plot2DFnc(fncBeale, xleft=-4.5, xright=4.5, ybottom=-4.5, ytop=4.5, 
          levels=[0.5, 5, 15, 50, 150, 500, 1000, 2500.0, 5000.0], mpts=mpt)
````
