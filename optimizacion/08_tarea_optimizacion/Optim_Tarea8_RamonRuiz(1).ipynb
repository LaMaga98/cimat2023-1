   En cada caso imprima los siguientes datos:
   
- la dimensión $n$,
- $f(\mathbf{x}_0)$,

**Función de cuadrática 1:** Para $\mathbf{x}=(x_1,x_2, ..., x_n)$

- $f(\mathbf{x}) = \frac{1}{2} \mathbf{x}^\top\mathbf{A}_1\mathbf{x} - \mathbf{b}_1^\top\mathbf{x}$,
  donde $\mathbf{A}_1$ y $\mathbf{b}_1$ están definidas por
  

$$ \mathbf{A}_1 = n\mathbf{I} + \mathbf{1} = 
\left[\begin{array}{llll} n      & 0      & \cdots & 0 \\
                       0      & n      & \cdots & 0 \\ 
                       \vdots & \vdots & \ddots & \vdots \\
                       0      & 0      & \cdots & n \end{array}\right]
+ \left[\begin{array}{llll} 1    & 1      & \cdots & 1 \\
                       1      & 1      & \cdots & 1 \\ 
                       \vdots & \vdots & \ddots & \vdots \\
                       1      & 1      & \cdots & 1 \end{array}\right],  \qquad
\mathbf{b}_1 = \left[\begin{array}{l} 1 \\ 1 \\ \vdots \\ 1 \end{array}\right], $$

donde $\mathbf{I}$ es la matriz identidad y $\mathbf{1}$ es la matriz llena de 1's,
ambas de tamaño $n$, usando los puntos iniciales   
- $\mathbf{x}_0 = (0,...,0)\in \mathbb{R}^{10}$ 
- $\mathbf{x}_0 = (0,...,0)\in \mathbb{R}^{100}$ 
- $\mathbf{x}_0 = (0,...,0)\in \mathbb{R}^{1000}$

  donde $\mathbf{A}_2= [a_{ij}]$ y $\mathbf{b}_2$ están definidas por
  
$$ a_{ij} = exp\left(-0.25(i-j)^2 \right),  \qquad
\mathbf{b}_2 = \left[\begin{array}{l} 1 \\ 1 \\ \vdots \\ 1 \end{array}\right] $$

**Función de Beale :** Para $\mathbf{x}=(x_1,x_2)$

$$f(\mathbf{x}) = (1.5-x_1 + x_1x_2)^2 + (2.25 - x_1 + x_1x_2^2)^2 + (2.625 - x_1 + x_1x_2^3)^2.$$

**Función de Himmelblau:** Para $\mathbf{x}=(x_1,x_2)$

- $\mathbf{x}_0 = (2,4)$

**Función de Rosenbrock:** Para $\mathbf{x}=(x_1,x_2, ..., x_n)$

$$ f(\mathbf{x}) = \sum_{i=1}^{n-1} \left[100(x_{i+1} - x_i^2)^2 + (1-x_i)^2 \right]
\quad n\geq 2.$$
- $\mathbf{x}_0 = (-1.2, 1.0)\in \mathbb{R}^{2}$  
- $\mathbf{x}_0 = (-1.2, 1.0, ..., -1.2, 1.0) \in \mathbb{R}^{20}$

```python
def backtracking(alpha_ini, xk, f, grad_f, p, rho, c1, Nb):
    alpha = alpha_ini
    t = rho*np.dot(grad_f(xk), p)
    k = 0
    while f(xk + alpha*p) > f(xk) + c1*alpha*t and k < Nb:
        alpha *= rho
        k += 1
    return alpha

def BFGS_mod(f, grad_f, x0, tol, H0, N, rho, c1, Nb):
    xk = x0
    gk = grad_f(xk) 
    Hk = H0
    n = len(xk)
    I = np.identity(n)
    for k in range(N):
        
        if np.linalg.norm(gk) <= tol:
            return xk, k, f(xk), np.linalg.norm(gk), True
            
        pk = np.dot(-Hk, gk)
        if np.dot(pk,gk) > 0:
            l1 = 10e-5 + (np.dot(pk,gk)/np.dot(gk,gk))
            Hk = Hk + l1*I
            pk = pk - l1*gk
        
        alpha = backtracking(1, xk, f, grad_f, pk, rho, c1, Nb)
        xk_new = xk + alpha*pk
        gk_new = grad_f(xk_new)
        y = gk_new - gk
        s = xk_new - xk

        if np.dot(y,y) < tol:
            return xk, k, f(xk), np.linalg.norm(gk), True

        if np.dot(y,s) <= 0:
            l2 = 10e-5 - (np.dot(y,s)/np.dot(y,y))
            Hk = Hk + l2*I
        else:
            rho_k = 1/np.dot(y,s)
            Hk = (np.identity(n) - rho_k * (s@y.T)) @ Hk @ (np.identity(n) - rho_k * (y@s.T)) + rho_k * (s@s.T)
        
        xk = xk_new
        gk = gk_new
        

    return xk, k, f(xk), np.linalg.norm(gk), False
```

```python
def Himmelblau(x):

    return (x[0]**2+x[1]-11)**2 + (x[0]+x[1]**2-7)**2

def grad_Himmelblau(x):

    d1=4*x[0]**3 + 4*x[0]*x[1] - 42*x[0] + 2*x[1]**2 - 14
    d2=4*x[1]**3 + 4*x[0]*x[1] + 2*x[0]**2 - 26*x[1] - 22

    return np.array([d1,d2])

def hess_Himmelbleau(x):
    d1=12*(x[0]**2) +4*x[1] -42
    d2=4*(x[0]+x[1])
    d3=4*x[0] +12*(x[1]**2) -25

    return np.array([[d1, d2], [d2, d3]])


def Beale(x):
    x1=x[0]
    x2=x[1]
    return (1.5 - x1 + x1*x2)**2 + (2.25 - x1 + x1*(x2**2))**2 + (2.625 - x1 + x1*(x2**3))**2

def grad_Beale(x):
    x1=x[0]
    x2=x[1]    
    dx1 = 2*(1.5 - x1 + x1*x2)*(-1 + x2) + 2*(2.25 - x1 + x1*(x2**2))*(-1 + x2**2) + 2*(2.625 - x1 + x1*(x2**3))*(-1 + x2**3)
    dx2 = 2*(1.5 - (x[0]) + (x[0])*(x[1]))*((x[0])) + 2*(2.25 - (x[0]) + (x[0])*(x[1])**2)*(2*(x[0])*(x[1])) + 2*(2.625 - (x[0]) + (x[0])*(x[1])**3)*(3*(x[0])*(x[1])**2)
    return np.array([dx1, dx2])

def hess_Beale(x):
    x1=x[0]
    x2=x[1] 
    d_x1x1 = 2 * (x2**6 + x2**4 - 2*x2**3 - x2**2 - 2*x2 + 3)
    d_x2x2 = x1 * (31.5*x2 + x1*(30*x2**4 + 3*x2**2 - 2*x2 - 2) + 9)
    d_x1x2 = 15.75*x2**2 + 9*x2 + 4*x1*(3*x2**5 + 2*x2**3 - 3*x2**2 - x2 - 1) + 3
    return [[d_x1x1, d_x1x2], [d_x1x2, d_x2x2]]



def Rosenbrock(x):
    n = len(x)
    suma = 0
    for i in range(n-1):
        suma += 100 * (x[i+1] - x[i]**2)**2 + (1 - x[i])**2
    return suma

def grad_Rosenbrock(x):
    n = len(x)
    gradient = np.zeros(n)
    for i in range(n-1):
        gradient[i] += -400 * x[i] * (x[i+1] - x[i]**2) - 2 * (1 - x[i])
        gradient[i+1] += 200 * (x[i+1] - x[i]**2)
    return gradient


def hess_Rosenbrock(x):
    n = len(x)
    hessian = np.zeros((n, n))
    for i in range(n - 1):
        hessian[i, i] += 1200 * x[i]**2 - 400 * x[i + 1] + 2
        hessian[i, i + 1] += -400 * x[i]
        hessian[i + 1, i] += -400 * x[i]
        hessian[i + 1, i + 1] += 200
    return hessian

def A1(n):
    return np.ones((n,n))+n*np.eye(n)

def A2(n):
    A = np.empty([n,n], dtype=float)
    for i in range(n):
        for j in range(n):
            u=0.25*((i-j)**2)
            A[i][j]=np.exp(-1*u)
    return A

def cuadratica(x):
    n = 10
    A = A1(n)
    b = np.ones(n)

    return 0.5 * np.dot(x.T, np.dot(A, x)) - np.dot(b.T, x)

def grad_cuadratica(x):
    n = 10
    A = A1(n)
    b = np.ones(n)
    return 0.5 * (np.dot(A.T, x) + np.dot(A, x)) - b


def hess_cuadratica(x):
    n = 10
    A = A1(n)

    return A
```
