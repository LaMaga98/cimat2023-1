### 3.Nos quedamos con los url que tengan el patron:

$<a href="(.*?)" class="image-post item-overlay " >$

pues son los que contienen la version estenográfica

```python
# Nombre del archivo .txt de entrada
archivo_entrada = "urls_conferencias.txt"

# Nombre del archivo de salida donde se guardarán los URLs extraídos
archivo_salida = "urls_extraidos.txt"

# Patrón de regex para encontrar URLs entre "<a href=" y " class="image-post item-overlay " >"
patron = r'<a href="(.*?)" class="image-post item-overlay " >'

# Abre el archivo de entrada en modo lectura
with open(archivo_entrada, "r") as f_entrada:
    # Lee el contenido del archivo
    contenido = f_entrada.read()

# Busca todas las coincidencias del patrón en el contenido del archivo
urls_extraidos = re.findall(patron, contenido)

# Abre el archivo de salida en modo escritura y guarda los URLs extraídos
with open(archivo_salida, "w") as f_salida:
    for url in urls_extraidos:
        f_salida.write(url + "\n")

print("Los URLs extraídos se han guardado en", archivo_salida)
```

### 5. Limpiamos el formato HTML

```python
#limpiamos
for j in range(i):    
    #archivo .txt con formato HTML
    ruta_inicio= f"raw_data/url_{j}.txt"
    # archivo de salida
    ruta_fin= f"clean_data/url_{j}.txt"

    #Leemos el contenido del archivo HTML
    with open(ruta_inicio, "r", encoding="utf-8") as archivo_html:
        contenido_html = archivo_html.read()

    #Beautiful Soup
    soup = BeautifulSoup(contenido_html, 'html.parser')

    # Eliminamos todas las etiquetas de script y estilo
    for script in soup(["script", "style"]):
        script.decompose()
    #texto limpio
    texto_limpio = soup.get_text()

    # Guardamps el texto limpio
    with open(ruta_fin, "w", encoding="utf-8") as archivo_salida:
        archivo_salida.write(texto_limpio)
```

### 6. Renombramos con la fecha correspondente y eliminamos aquellos archivos que no contengan la fecha 

(pues estos no contienen información)

```python
# Directorio donde se encuentran los archivos .txt
directorio = "clean_data"

# Patrón de regex para buscar la fecha en formato DD.MM.AA
patron_fecha = r'\b\d{2}\.\d{2}\.\d{2}\b'

# Diccionario para rastrear la cantidad de veces que se ha encontrado cada fecha
conteo_fechas = {}

# Itera sobre cada archivo en el directorio
for nombre_archivo in os.listdir(directorio):
    ruta_archivo = os.path.join(directorio, nombre_archivo)
    
    # Verifica si el archivo es un archivo .txt
    if nombre_archivo.endswith('.txt') and os.path.isfile(ruta_archivo):
        # Abre el archivo en modo lectura
        with open(ruta_archivo, 'r') as archivo:
            # Lee el contenido completo del archivo
            contenido = archivo.read()
            
            # Busca todas las coincidencias de fecha en el contenido usando regex
            fechas_encontradas = re.findall(patron_fecha, contenido)
            
            # Si se encontraron fechas
            if fechas_encontradas:
                for fecha in fechas_encontradas:
                    # Formatea la fecha como DD-MM-AA
                    fecha_formateada = fecha.replace('.', '-')
                    
                    # Si la fecha ya ha sido encontrada antes, incrementa el contador
                    if fecha_formateada in conteo_fechas:
                        conteo_fechas[fecha_formateada] += 1
                        nuevo_nombre = f"{fecha_formateada}_{conteo_fechas[fecha_formateada]}.txt"
                    else:
                        # Si es la primera vez que se encuentra la fecha, inicializa el contador
                        conteo_fechas[fecha_formateada] = 1
                        nuevo_nombre = f"{fecha_formateada}.txt"
                    
                    # Renombra el archivo con el nuevo nombre
                    os.rename(ruta_archivo, os.path.join(directorio, nuevo_nombre))
                    print(f"Archivo renombrado: {nombre_archivo} -> {nuevo_nombre}")
                    # Rompe el bucle después de renombrar el archivo para pasar al siguiente archivo
                    break
            else:
                print(f"No se encontró una fecha en el archivo: {nombre_archivo}")
                # Elimina el archivo donde no se encontró ninguna fecha
                print(f"Archivo eliminado: {nombre_archivo}")
                os.remove(ruta_archivo)
                
    else:
        print(f"El archivo {nombre_archivo} no es un archivo .txt o no existe.")
```
