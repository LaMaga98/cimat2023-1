5. Creamos una única base de datos con los precios de cada producto de acuerdo al mes, año y ciudad

```python
dfs = [limon, leche, camaron, tortillas, electricidad_225kw, alquiler, internet, frijol, pollo]
#dataframe vacío para almacenar el resultado combinado
base_productos = pd.DataFrame()

#Iteramos sobre los dataframes y combinarlos en uno solo
for producto, df in zip(productos, dfs):
    # Renombramos las columnas
    df.rename(columns={'Precio promedio': producto}, inplace=True)
    if base_productos.empty:
        base_productos = df
    else:
        base_productos = pd.merge(base_productos, df, on=['A�o', 'Mes', 'Clave ciudad'], how='outer')
```

```python
#Renombramos la columna Año
base_productos.rename(columns={'A�o': 'Anio'}, inplace=True)
#Exportamos el dataframe a un archivo CSV
base_productos.to_csv(carpeta+'base_productos.csv', index=False)
```
