$\color{lightblue}{Inciso \space c}$ <br>

```python
def loss_function(true_label, predicted_label):
    if true_label != predicted_label:
        return 1
    else:
        return 0
    
def naive_bayes_classifier(X_train, y_train, X_test, loss):
    predicted_classes = []

    # Valores únicos que toma Y
    etiquetas = y_train.unique()
    # Número total de observaciones en y_train
    y_total = len(y_train)

    # Probabilidad de que Y=0 es número_de_veces_Y=0/total_observaciones
    p_y0 = y_train.value_counts()[etiquetas[0]] / y_total 
    # Probabilidad de que Y=1 es número_de_veces_Y=1/total_observaciones
    p_y1 = y_train.value_counts()[etiquetas[1]] / y_total 

    # Calculamos c1
    c1 = (loss(0, 1) * p_y0) / (loss(1, 0) * p_y1)

    #Calculamos promedio y varianza de todas las caracteristicas POR VALOR DE Y

    #mascaras para valor de Y
    etiqueta0=(y_train == y_train.unique()[0])
    etiqueta1=(y_train == y_train.unique()[1])
    #encontramos los indices correspondientes
    indice0 = np.where(etiqueta0)
    indice1= np.where(etiqueta1)

    mu_y0=[ X_train[i].iloc[indice0].mean() for i in X_test]
    mu_y1=[ X_train[i].iloc[indice1].mean() for i in X_test]
    var_y0=[ X_train[i].iloc[indice0].var() for i in X_test]
    var_y1=[ X_train[i].iloc[indice1].var() for i in X_test]

        

    for j in range(len(X_test)):

        for k,i in enumerate(X_test):
            #inicializamos la probailidad en 1
            prob_y0 = 1
            prob_y1 = 1

            #Calculamos la probabilidad decada xj usando que son X|Y=y son independientes y que 
            #tienen distribucion normal univariada. 
            #Para ello usamos los promedios y varianzas calculados antes

            #Calculamos la probabilidad condicional para Y=0
            prob_y0 *= (1 / np.sqrt(2 * np.pi * var_y0[k])) * np.exp(-((X_test.iloc[j,k] - mu_y0[k]) ** 2) / (2 * var_y0[k]))
            # Calculamos la probabilidad condicional para Y=1
            prob_y1 *= (1 / np.sqrt(2 * np.pi * var_y1[k])) * np.exp(-((X_test.iloc[j,k] - mu_y1[k]) ** 2) / (2 * var_y1[k]))
    
        # Clasificador bayesiano
        if c1 * prob_y1 > prob_y0:
            predicted_classes.append(etiquetas[1])
        else:
            predicted_classes.append(etiquetas[0])

    return predicted_classes

```
